```python
## matplotlib exercises continued
## Exercise 5

import matplotlib.pyplot as plt

import math


x = range(1, 100)
y1 = [((0 * n) + 1) for n in x]
y2 = [(math.log(n)) for n in x]
y3 = [n for n in x]
y4 = [(n * math.log(n)) for n in x]
y5 = [(n ** 2) for n in x]
y6 = [(2 ** n) for n in x]
y7 = [(math.factorial(n)) for n in x]
y8 = [(n ** n) for n in x]

plt.plot(x,y1, c='g', label='0(n) + 1')
plt.plot(x,y2, c='b', label='log(n)')
plt.plot(x,y3, c='y', label='n')
plt.plot(x,y4, c='c', label='n * log(n)')
plt.plot(x,y5, c='r', label='$n^2$')
plt.plot(x,y6, c='k', label='$2^n$')
plt.plot(x,y7, c='m', label='n!')
plt.plot(x,y8, c='darkorange', label='$n^n$')

plt.title('Big O Notation')
plt.legend(loc= 'upper right')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.xlim(0,20)
plt.ylim(-1,50)

plt.show()
```
