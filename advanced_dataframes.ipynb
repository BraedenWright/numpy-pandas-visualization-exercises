```python
#  1. pull Chipotle database from MySQL

chip_db = f'mysql+pymysql://{user}:{password}@{host}/chipotle'

chipotle = pd.read_sql('''SELECT *
        FROM orders''', chip_db)

chipotle.head()
```

```python
#  analyze the data

chipotle.info()
```

```python
chipotle.describe()
```

```python
#   2. whats the total price for each order

chipotle['item_price'] = chipotle.item_price.apply(lambda value: float(value[1:].strip()))


chipotle.groupby(['order_id']).item_price.agg('sum')
```

```python
#   3. the three most ordered items

chipotle.groupby('item_name').quantity.agg('sum').sort_values().tail(3)
```

```python
#   4.  Which item produces the most revenue?

chipotle.groupby('item_name').item_price.agg('sum').sort_values().tail(1)
```

```python
#   5.  Join employees and tittles dataframes together

employees_and_titles = pd.merge(employees_db, titles_db, left_on='emp_no', right_on='emp_no', how='inner')

employees_and_titles
```

```python
#   6. FOR EACH TITLE, FIND THE HIRE DATE OF THE MOST RECENTLY EMPLOYEE OF EACH TITLE
#      might be easier to filter with a query


recent_hire = pd.read_sql('''
                SELECT title, max(from_date)
                AS 'recent_hire'
                FROM employees
                JOIN titles USING(emp_no)
                GROUP BY TITLE;''', get_db_url)

recent_hire
```

```python
#  7.  Cross Tabulate title by department

#pd.crosstab(titles_db.title)

# gonna need to join the dept_emp table to find titles in each dept
#  also make sure to limit

title_in_dept = pd.read_sql('''SELECT title, dept_name
            FROM employees
            JOIN titles USING(emp_no)
            JOIN dept_emp USING(emp_no)
            JOIN departments USING(dept_no)
            WHERE titles.to_date = "9999-01-01" ''', get_db_url)

title_in_dept
```
