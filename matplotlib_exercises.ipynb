```python
import matplotlib.pyplot as plt

print(plt)
```

```python
## Exercise 1

x = range(-50, 50)
y = [(x**2 - x + 2) for x in x]

plt.plot(x, y, c='b')
plt.title('Exercise 1, $y = x^2 - x + 2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
## Exercise 2

import math


x = range(0, 100)
y = [(math.sqrt(x)) for x in x]

plt.plot(x,y, c='r')
plt.title('$y = √x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(0,100)
plt.ylim(0,10)
plt.show()
```

```python
x = range(0, 100)
y = [(x ** 3) for x in x]

plt.plot(x,y, c='g')
plt.title('$y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

```python
x = range(0, 15)
y = [(2 ** x) for x in x]

plt.plot(x,y, c='b')
plt.title('$y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
x = range(0, 15)
y = [(1 / (x + 1)) for x in x]

plt.plot(x,y, c='y')
plt.title('$y = 1 / (x + 1)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
## Exercise 3
## put all my graphs in one figure with some cool details

plt.figure(figsize=(12, 8))

plt.subplot(221)

x = range(0, 100)
y = [(math.sqrt(x)) for x in x]

plt.plot(x,y, c='r')
plt.title('$y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(0,100)
plt.ylim(0,10)

plt.subplot(222)

x = range(0, 15)
y = [(2 ** x) for x in x]

plt.plot(x,y, c='b')
plt.title('$y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(223)

x = range(0, 100)
y = [(x ** 3) for x in x]

plt.plot(x,y, c='g')
plt.title('$y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.subplot(224)

x = range(0, 15)
y = [(1 / (x + 1)) for x in x]

plt.plot(x,y, c='y')
plt.title('$y = 1 / (x + 1)$')
plt.xlabel('$x$')
plt.ylabel('$y$')


plt.show()
```

```python
x = range(0, 50)
y1 = [(math.sqrt(x)) for x in x]
y2 = [(2 ** x) for x in x]
y3 = [(x ** 3) for x in x]
y4 = [(1 / (x + 1)) for x in x]

plt.plot(x,y1, c='r', marker ='v')
plt.plot(x,y2, c='b')
plt.plot(x,y3, c='g')
plt.plot(x,y4, c='y')

plt.show()
```
